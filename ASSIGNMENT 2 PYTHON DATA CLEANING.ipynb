```python
del df["Id"]
```

```python
def extract_name(email):
    if pd.isnull(email):
        return email
    else:
        return email.split("@")[0]
```

```python
df['Name'] = df.apply(lambda row: extract_name(row['Email']) if pd.isnull(row['Name']) else row['Name'], axis=1)
```

```python
mean_age = df['Age'].mean()
```

```python
df['Age'].fillna(mean_age, inplace=True)
```

```python
df.dropna(subset=['Name', 'Email'], how='all', inplace=True)
```

```python
df.loc[df['Email'].isnull() & df['Name'].notnull(), 'Email'] = 'support@dataisgood.com'
```

```python
df['Address'].fillna('(Address not Available)', inplace=True)
```

```python
df_sorted = df.sort_values('Salary', ascending=False)
df_sorted['Top 10'] = False
df_sorted.loc[:9, 'Top 10'] = True
highlight = df_sorted['Top 10'].map({True: 'background-color: yellow', False: ''})
df_sorted.style.apply(lambda _: highlight)
```
